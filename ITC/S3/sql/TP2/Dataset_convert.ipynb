```python
import pandas as pd
```

```python
frame = pd.read_csv("fr-esr-sise-effectifs-d-etudiants-inscrits-esr-public.csv", on_bad_lines="skip", sep=";")
```

```python
pd.set_option('display.max_columns', None)
frame
```

```python
eleves = frame[["etablissement_id_paysage", "SEXE", "Série de baccalauréat", "Attractivité départementale (département d'obtention du baccalauréat)", "Diplôme", "Secteur disciplinaire", "Nombre total d'étudiants inscrits (inscriptions principales et secondes) hors étudiants inscrits en parallèle en CPGE", "Nombre total d'étudiants inscrits (inscriptions principales et secondes) y compris étudiants inscrits en parallèle en CPGE"]]
```

```python
eleves = eleves.rename(columns={
"etablissement_id_paysage": "id_etablissement",
"SEXE": "sexe",
"Série de baccalauréat": "serie_bac",
"Attractivité départementale (département d'obtention du baccalauréat)": "departement_bac", 
"Diplôme": "diplome",
"Secteur disciplinaire": "discipline",
"Nombre total d'étudiants inscrits (inscriptions principales et secondes) hors étudiants inscrits en parallèle en CPGE": "nb_etudiants_sans_CPGE", 
"Nombre total d'étudiants inscrits (inscriptions principales et secondes) y compris étudiants inscrits en parallèle en CPGE": "nb_etudiants_avec_CPGE" 
})
```

```python
eleves = eleves.groupby(by=["id_etablissement", "sexe", "serie_bac", "departement_bac", "diplome", "discipline"]).sum().reset_index().set_index([ "id_etablissement",
"sexe",
"serie_bac",
"departement_bac", 
"diplome",
"discipline"])
```

```python
eleves
```

```python
etablissements = frame[["etablissement_id_paysage", "Type d'établissement", "Etablissement", "etablissement_code_commune"]].drop_duplicates().reset_index(drop=True)
```

```python
etablissements = etablissements.rename(columns={
    "Type d'établissement": "type", 
    "etablissement_id_paysage": "id", 
    "Etablissement": "nom",
    "etablissement_code_commune": "code_commune"
})
etablissements = etablissements.set_index("id")
```

```python
etablissements
```

```python
communes = frame[["etablissement_code_commune", "Commune du siège de l’établissement", "Département du siège de l’établissement", "Région du siège de l’établissement"]].drop_duplicates().reset_index(drop=True)
```

```python
communes = communes.rename(columns={
    "etablissement_code_commune": "code", 
    "Commune du siège de l’établissement": "nom",
    "Département du siège de l’établissement": "departement",
    "Région du siège de l’établissement": "region"
})
communes = communes.set_index("code")
```

```python
communes
```

```python
from sqlalchemy import create_engine

engine = create_engine('sqlite:///EnseignementSup.db', echo=False)
```

```python
eleves.to_sql(name='Eleves', con=engine)
etablissements.to_sql(name='Etablissements', con=engine)
communes.to_sql(name='Communes', con=engine)
```

```python
from sqlalchemy import text

with engine.connect() as conn:
   print(conn.execute(text("SELECT * FROM Communes")).fetchall())
```
